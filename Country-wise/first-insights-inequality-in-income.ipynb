```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import random

import warnings as wrn
wrn.filterwarnings("ignore")

df = pd.read_csv("Inequality in Income.csv",encoding="utf8")
```

```python
def color_generator():
    while True:
        
        r = random.randint(30,255)/255
        g = random.randint(30,255)/255
        b = random.randint(30,255)/255
        a = random.randint(100,255)/255
        yield (r,g,b,a)
```

```python
color_gen = color_generator()
```

## Distribution of Features
<a id="2"/>

### 1. Distribution of Categorical Features
<a id="3"/>

```python
categorical_features = ["Continent","Hemisphere","Human Development Groups"]

fig,ax = plt.subplots(ncols=3,figsize=(14,4))
for i in range(len(categorical_features)):
    
    names = list(df[categorical_features[i]].value_counts().to_dict().keys())
    values = list(df[categorical_features[i]].value_counts().to_dict().values())
    ax[i].bar(names,values,color=[next(color_gen) for i in range(len(names))])
    ax[i].set_title(categorical_features[i])
    ax[i].tick_params(axis="x",rotation=45)
    
fig.suptitle("Distribution of Categorical Features",fontsize=13)
plt.show()
```

* Most countries listed in the dataset is in Northern Hemisphere.

### 2. Distribution of Numerical Features
<a id="4"/>

```python
# 12 years
numerical_features = df.columns[7:].to_list()

fig,ax = plt.subplots(ncols=4,nrows=3,figsize=(12,12))

i = 0
for vertical_ax in ax:
    for column_ax in vertical_ax: 
        column_ax.hist(df[numerical_features[i]],color=next(color_gen))
        column_ax.set_title(numerical_features[i])
        i += 1
        
```

## The World In General 
<a id="5"/>

* **How did inequality in income change over the years in the world? (each years' median value)**
<a id="6"/>

```python
median_inequality_world = df.describe().loc["50%"][1:].to_list()
years = range(2010,2022)


figure = plt.figure(figsize=(14,7))
plt.plot(years,median_inequality_world)
plt.title("Median Inequality in Income Over The Years")
plt.show()
```

* **How did inequality in income change by continents over the years?**
<a id="7"/>

```python
income_columns = df.columns[7:]
years = range(2010,2022)
labels = ["Africa","America","Asia","Europe","Ocenia"]

# 12 years 5 continents
values = np.zeros((12,5))

for i in range(len(income_columns)):
    values[i] = df.groupby("Continent")[income_columns[i]].describe()["50%"].to_numpy()
    

plt.figure(figsize=(10,4))
for c_value in values.T:
    plt.plot(years,c_value)

plt.legend(labels,loc="upper right")
plt.show()
```

## 10 Most Developed Countries By Their HDI
<a id="8"/>

In this section, we're going to take a look at most developed ten countries according to their **HDI (Human Development Index) ranks.**

* **In which continent are those countries?**
<a id="9"/>

```python
top_ten_countries = df.sort_values("HDI Rank (2021)")[:10]
```

```python
names = list(top_ten_countries["Continent"].value_counts().to_dict().keys())
values = list(top_ten_countries["Continent"].value_counts().to_dict().values())

plt.figure(figsize=(6,3))
plt.title("Continents of 10 Most Developed Countries")
plt.bar(names,values,color=[next(color_gen) for i in range(len(names))])
plt.show()
```

* **How did inequality in income change over the years in those countries?**
<a id="10"/>

```python
years = range(2010,2022)

figure = plt.figure(figsize=(14,7))
for country in top_ten_countries.iloc:
    plt.plot(years,country.iloc[7:].to_numpy(),label=country["Country"])
    plt.legend(loc="upper right")
    
plt.title("Inequality In Income Over The Years\n\n10 Most Developed Countries")
plt.show()
```

## 10 Least Developed Countries By Their HDI

In this section, we're going to take a look at ten least developed countries according to their **HDI (Human Development Index) ranks.**
<a id="11"/>

* **In which continent are those countries?**
<a id="12"/>

```python
# We had to drop null values because some countries don't have their HDI information.
least_ten_countries = df.sort_values("HDI Rank (2021)").dropna().iloc[-10:]


names = list(least_ten_countries["Continent"].value_counts().to_dict().keys())
values = list(least_ten_countries["Continent"].value_counts().to_dict().values())

plt.figure(figsize=(6,3))
plt.title("Continents of 10 Least Developed Countries")
plt.bar(names,values,color=[next(color_gen) for i in range(len(names))])
plt.show()
```

* **How did inequality in income change over the years in those countries?**
<a id="13"/>

```python
years = range(2010,2022)

figure = plt.figure(figsize=(14,7))
for country in least_ten_countries.iloc:
    plt.plot(years,country.iloc[7:].to_numpy(),label=country["Country"])
    plt.legend(loc="upper right")
    
plt.title("Inequality In Income Over The Years\n\n10 Least Developed Countries")
plt.show()
```
